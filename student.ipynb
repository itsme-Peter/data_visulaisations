```python
# Your code here - remember to use markdown cells for comments as well!

#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#defining dataset
train_csv = pd.read_csv("./train.csv")
train_csv = train_csv[["SalePrice", "TotRmsAbvGrd", "OverallCond"]]
print(train_csv.describe())
```

```python
# Create line plots
plt.figure(figsize=(15, 5))

# Line plot for SalePrice
plt.plot(train_csv['SalePrice'], label='SalePrice')
plt.show()

# Line plot for TotRmsAbvGrd
plt.plot(train_csv['TotRmsAbvGrd'], label='TotRmsAbvGrd')
plt.show()

# # Line plot for OverallCond
plt.plot(train_csv['OverallCond'], label='OverallCond')
plt.show()

plt.xlabel('Id')
plt.ylabel('Values')
plt.title('Line Graph of SalePrice, TotRmsAbvGrd, and OverallCond')
plt.legend()
plt.show()
```

```python
# Histograms visualizations
#define the figure

plt.figure(figsize=(15,10))

plt.subplot(3, 2, 1)
sns.histplot(train_csv['SalePrice'], kde=True)
plt.title('SalePrice Distribution')

plt.subplot(3, 2, 2)
sns.histplot(train_csv['TotRmsAbvGrd'], kde=True)
plt.title('TotRmsAbvGrd Distribution')

plt.show()

plt.subplot()
sns.histplot(train_csv['OverallCond'], kde=True)
plt.title('OverallCond Distribution')

plt.show()
```

```python
# Scatter plots
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(x=train_csv['TotRmsAbvGrd'], y=train_csv['SalePrice'])
plt.title('SalePrice vs TotRmsAbvGrd')

plt.subplot(1, 2, 2)
sns.scatterplot(x=train_csv['OverallCond'], y=train_csv['SalePrice'])
plt.title('SalePrice vs OverallCond')

plt.tight_layout()
plt.show()
```

```python
# Line plots for each variable
plt.subplot(3, 1, 1)
sns.lineplot(data=train_csv, x='YrSold', y='SalePrice')
plt.title('Average SalePrice by Year')
plt.xlabel('Year')
plt.ylabel('Average SalePrice')

plt.subplot(3, 1, 2)
sns.lineplot(data=train_csv, x='YrSold', y='TotRmsAbvGrd')
plt.title('Average TotRmsAbvGrd by Year')
plt.xlabel('Year')
plt.ylabel('Average TotRmsAbvGrd')

plt.subplot(3, 1, 3)
sns.lineplot(data=train_csv, x='YrSold', y='OverallCond')
plt.title('Average OverallCond by Year')
plt.xlabel('Year')
plt.ylabel('Average OverallCond')

plt.tight_layout()
plt.show()
```
